### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_start, x_end = 0, 10

if not (0 <= x_start < x_end <= 100):
    raise ValueError("x values must be in the range 0 to 100 and start < end.")

x = np.arange(x_start, x_end + 1)
y = x ** 2

plt.plot(x, y, marker='o', color='blue')
plt.title("Square of x from 0 to 10")
plt.xlabel("x values")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```
