### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

x_line = [1, 2, 3, 4]
y_line = [1, 4, 9, 16]

categories_bar = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 6]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 1, 8, 7]

labels_pie = ['Python', 'Java', 'C++', 'JS']
sizes_pie = [30, 25, 25, 20]

if len(x_line) != len(y_line):
    raise ValueError("Line plot: x and y must be the same length.")
if len(categories_bar) != len(values_bar):
    raise ValueError("Bar plot: categories and values must be the same length.")
if len(x_scatter) != len(y_scatter):
    raise ValueError("Scatter plot: x and y must be the same length.")
if sum(sizes_pie) != 100:
    raise ValueError("Pie chart: sizes must sum to 100.")

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x_line, y_line, marker='o')
axs[0, 0].set_title("Line Chart")

axs[0, 1].bar(categories_bar, values_bar, color='skyblue')
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title("Scatter Plot")

axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
axs[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
